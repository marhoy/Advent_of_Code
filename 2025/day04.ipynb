```python
from pathlib import Path

file = Path("day04_input.txt")

grid = {}
for row, line in enumerate(file.read_text().splitlines()):
    for col, char in enumerate(line):
        grid[(row, col)] = char
```

```python
def count_adjacent_rolls(grid):
    """Count the number of adjacent '@' characters for each position in the grid."""
    adjacent_counts = {}
    for row, col in grid:
        count = 0
        for dr in [-1, 0, 1]:
            for dc in [-1, 0, 1]:
                if dr == 0 and dc == 0:
                    continue
                neighbor = (row + dr, col + dc)
                if grid.get(neighbor) == "@":
                    count += 1
        adjacent_counts[(row, col)] = count
    return adjacent_counts
```

```python
def can_be_accessed(grid, adjacent_counts):
    """Find positions that can be accessed based on adjacent '@' counts."""
    return [
        pos for pos, count in adjacent_counts.items() if grid[pos] == "@" and count < 4
    ]
```

# Part 1


```python
adjacent_counts = count_adjacent_rolls(grid)
len(can_be_accessed(grid, adjacent_counts))
```

# Part 2


```python
removed = 0
while True:
    adjacent_counts = count_adjacent_rolls(grid)
    accessible_positions = can_be_accessed(grid, adjacent_counts)
    if len(accessible_positions) == 0:
        break
    removed += len(accessible_positions)
    for pos in accessible_positions:
        grid[pos] = "."
    print(f"Removed {len(accessible_positions)} positions this round.")

print("Removed in total: ", removed)
```
