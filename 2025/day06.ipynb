```python
from functools import reduce
from operator import add, mul
from pathlib import Path

with Path("day06_input.txt").open() as f:
    lines = list(f)

ops = {"+": add, "*": mul}
```

# Part 1


```python
numbers = [[int(x) for x in line.split()] for line in lines[:-1]]
operators = lines[-1].split()
rows, cols = len(numbers), len(numbers[0])

result = 0
for col in range(cols):
    selection = [numbers[row][col] for row in range(rows)]
    operation = ops[operators[col]]
    result += reduce(operation, selection)

result
```

# Part 2


```python
cols = len(lines[0])
iter_ops = iter(ops[op] for op in lines[-1].split())

result = 0
numbers = []
for col in range(cols):
    col_chars = [line[col] for line in lines[:-1]]
    if all(c in (" ", "\n") for c in col_chars):
        # We've hit a separator column, perform the operation
        result += reduce(next(iter_ops), numbers)
        numbers = []
    else:
        # Add this number to the current problem
        numbers.append(int("".join(col_chars)))

result
```
