```python
from pathlib import Path

lines = Path("day07_input.txt").read_text().splitlines()
```

# Part 1


```python
beams = {lines[0].find("S")}
splits = 0
for row in lines[1:]:
    next_beams = beams.copy()
    for col, char in enumerate(row):
        if char == "^" and col in beams:
            splits += 1
            next_beams -= {col}
            next_beams |= {col - 1, col + 1}
    beams = next_beams

splits
```

# Part 2


```python
particles = [0] * len(lines[0])
particles[lines[0].find("S")] = 1

for row in lines[1:]:
    next_particles = particles.copy()
    for col, char in enumerate(row):
        if char == "^" and particles[col] > 0:
            next_particles[col] = 0
            next_particles[col - 1] += particles[col]
            next_particles[col + 1] += particles[col]
    particles = next_particles

sum(particles)
```
