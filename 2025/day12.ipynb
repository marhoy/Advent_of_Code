```python
import re
from pathlib import Path

from pydantic import BaseModel
```

```python
file_contents = Path("day12_input.txt").read_text()
```

```python
shape_area = {}
for line in file_contents.split("\n\n"):
    if match := re.match(r"^(\d+):", line):
        shape_area[int(match[1])] = line.count("#")
shape_area
```

```python
class Region(BaseModel):
    """The region where shapes can be placed."""

    area: int
    shape_numbers: list[int]


regions: list[Region] = []
for line in file_contents.split("\n"):
    if match := re.match(r"^(\d+)x(\d+):(.+)", line):
        regions.append(
            Region(
                area=int(match[1]) * int(match[2]),
                shape_numbers=[int(n) for n in match[3].strip().split(" ")],
            )
        )
```

# Part 1

Apparently, for the input it's enough to check that the sum of the package-areas would
fit into the area of the region. We don't have to worry about rotating, flipping, and
placing the packages on the grid...


```python
possible = 0
for region in regions:
    total_shape_area = sum(
        n * shape_area[i] for i, n in enumerate(region.shape_numbers)
    )
    if total_shape_area < region.area:
        possible += 1
possible
```
