```python
from pathlib import Path

banks = Path("day03_input.txt").read_text().strip().splitlines()
```

# Part 1


```python
total = 0
for bank in banks:
    first = max(bank[:-1])
    first_index = bank.index(first)
    second = max(bank[first_index + 1 :])
    total += int(first + second)

print(total)
```

# Part 2

Same idea as in part 1, only generalized to 12 digits.

- The first digit cannot be from the last 12 digits.
- The second digit must be after the first, but not among the last 11.
- etc.


```python
total = 0
for bank in banks:
    digits = []
    start = 0
    end = len(bank) - 12
    for _ in range(12):
        digit = max(bank[start : end + 1])
        digits.append(digit)
        start = bank.index(digit, start) + 1
        end += 1
    total += int("".join(digits))

print(total)
```
