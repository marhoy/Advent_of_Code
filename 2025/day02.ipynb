```python
from pathlib import Path
```

```python
input_file = Path("day02_input.txt")
ranges = input_file.read_text().strip().split(",")
```

```python
def is_invalid_id_part_1(id_str: str) -> bool:
    """Determine if the given ID string is invalid.

    An ID is invalid if it consists of two identical halves.
    """
    # If the length of the string is odd, it must be valid
    if len(id_str) % 2 != 0:
        return False
    # Split the string into two halves
    mid = len(id_str) // 2
    first_half = id_str[:mid]
    second_half = id_str[mid:]
    # If the first half equals the second half, the ID is invalid
    return first_half == second_half
```

# Part 1


```python
total = 0
for start_end in ranges:
    start, end = map(int, start_end.split("-"))
    for id_num in range(start, end + 1):
        if is_invalid_id_part_1(str(id_num)):
            total += id_num

print(f"Total of invalid IDs: {total}")
```

# Part 2


```python
def is_invalid_id_part_2(id_str: str) -> bool:
    """Determine if the given ID string is invalid.

    An ID is invalid if it is created by repeating a pattern at least twice.
    """
    length = len(id_str)
    # Check for all possible pattern lengths
    for pattern_length in range(1, length // 2 + 1):
        if length % pattern_length == 0:  # Pattern must fit evenly
            pattern = id_str[:pattern_length]
            repetitions = length // pattern_length
            if pattern * repetitions == id_str:
                return True
    return False
```

```python
total = 0
for start_end in ranges:
    start, end = map(int, start_end.split("-"))
    for id_num in range(start, end + 1):
        if is_invalid_id_part_2(str(id_num)):
            total += id_num

print(f"Total of invalid IDs: {total}")
```
