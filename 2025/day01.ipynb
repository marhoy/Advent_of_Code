```python
from pathlib import Path
```

```python
input_file = Path("day01_input.txt")
turns = input_file.read_text().splitlines()
```

# Part 1


```python
position = 50
num_positions = 100
num_zeros = 0

for turn in turns:
    direction, steps = turn[0], int(turn[1:])
    if direction == "L":
        steps = -steps
    position = (position + steps) % num_positions
    num_zeros += position == 0

num_zeros
```

# Part 2


```python
position = 50
num_positions = 100
num_zeros = 0

for turn in turns:
    direction, steps = turn[0], int(turn[1:])

    # First: Count number of full cycles
    num_zeros += steps // num_positions
    # Reduce steps to remainder after full cycles
    steps = steps % num_positions

    if direction == "L":
        steps = -steps

    # Check if we would pass position 0 when moving. It does not count if we start at 0.
    if position != 0 and not 0 <= position + steps <= 100:
        num_zeros += 1

    # Move to new position
    position = (position + steps) % num_positions

    # Check if we ended at position 0
    num_zeros += position == 0

num_zeros
```
