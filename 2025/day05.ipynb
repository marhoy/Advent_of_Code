```python
from pathlib import Path

file = Path("day05_input.txt")

ranges = []
ingredients = []

for line in file.open():
    if "-" in line:
        start, end = map(int, line.strip().split("-"))
        ranges.append((start, end))
    elif line.strip():
        ingredients.append(int(line.strip()))
```

# Part 1


```python
def is_fresh(ingredient, ranges):
    """Determine if an ingredient is fresh based on the given ranges."""
    return any(start <= ingredient <= end for start, end in ranges)
```

```python
sum(1 for ingredient in ingredients if is_fresh(ingredient, ranges))
```

# Part 2


```python
import portion as P  # noqa: N812
```

```python
# Merge overlapping ranges
intervals = P.empty()
for start, end in ranges:
    intervals |= P.closed(start, end)

# Calculate total length of merged intervals
total = 0
for interval in P.to_data(intervals):
    lower, upper = interval[1:3]
    total += upper - lower + 1

total
```
